```python
database                = db.Database(DB_CONFIG)
items_repository        = db.ItemRepository(database)
interactions_repository = db.InteractionRepository(database)
```

```python
items        = pd.read_json(f'{DATASET_PATH}/db_items.json')
interactions = pd.read_json(f'{DATASET_PATH}/db_interactions.json')
```

```python
database.reset()
```

```python
import re
# as per recommendation from @freylis, compile once only
CLEANR = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')

def cleanhtml(raw_html):  return re.sub(CLEANR, '', raw_html) if raw_html else '' 
```

```python
items['description'] = items['description'].apply(cleanhtml)
```

```python
items_repository.save_all(items)
```

```python
interactions_repository.save_all(interactions)
```
